```python
step = 1
inicio = [100, 0]
fim = [500, 1000]

divisor = 3

multdois = list()
multcinco = []
multsete = []
ldivisor = list()
maiusculo = list()
minusculo = list()

# variáveis do tipo string
nome = 'Joããão da Silva'
cidade = 'São Paulo'
cpf = '123.456.789-00'

```

```python
for numero in range(inicio[0], fim[0], step):
    if numero % 2 == 0:
        multdois.append(numero)
    if numero % 5 == 0:
        multcinco.append(numero)
    if numero % 7 == 0: 
        multsete.append(numero)
print(multdois)
print(multcinco)
print(multsete)
```

```python
for n in range(inicio[1], fim[1], step):
  if n % divisor == 0:
    ldivisor.append(n)
print(ldivisor)
```

```python
#função para converter caracter em maíuscula ou minúscula
#parametros: lista de strings, tipo "M" para maiúsculo e "m" para mínusculo
lista = [nome, cidade]
tipos = 'M'
def convertecaracter(palavras, tipo):
  try:
    result = []
    for p in palavras:
      if tipo == 'M': 
        result.append(p.upper())
      elif tipo =='m': 
        result.append(p.lower())
    return result
  except: 
    print('Ops! Algo aconteceu!, revise os parâmetros')  
print(convertecaracter(lista, tipos))
```
